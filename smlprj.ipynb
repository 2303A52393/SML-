```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Student_performance_data _ (1).csv')
df
```

```python
y=df['GradeClass']
y
```

```python
x=df.drop('GradeClass',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, y_pred)
print(confusion)
```

```python
from sklearn.utils import resample
l_acc=[]
s_acc=[]
k_acc=[]
for i in range(0,20):

  x_r,y_r=resample(x,y,n_samples=200)
  x_tr,x_te,y_tr,y_te=train_test_split(x_r,y_r,test_size=0.2,random_state=42)
  model=LogisticRegression()
  model.fit(x_tr,y_tr)
  y_pred=model.predict(x_te)
  accuracy=accuracy_score(y_te,y_pred)
  l_acc.append(accuracy)
  svc=SVC()
  svc.fit(x_tr,y_tr)
  y_p=svc.predict(x_te)
  accuracy=accuracy_score(y_te,y_p)
  s_acc.append(accuracy)
  knn=KNeighborsClassifier(n_neighbors=3)
  knn.fit(x_tr,y_tr)
  y_pr=knn.predict(x_te)
  accuracy=accuracy_score(y_te,y_pr)
  k_acc.append(accuracy)
```

```python
import matplotlib.pyplot as plt
plt.plot(k_acc,label="knn")
plt.plot(s_acc,label="svc")
plt.plot(l_acc,label="log")
plt.legend()
```
